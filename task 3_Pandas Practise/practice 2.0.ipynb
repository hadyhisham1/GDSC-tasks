# histplot|

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.__version__
```

```python
peng = sns.load_dataset("penguins")
```

```python
peng.head()
```

```python
peng.shape
```

```python
peng.dropna(inplace=True)
```

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(peng.bill_length_mm)
```

```python
sns.histplot(y='bill_length_mm',
             data=peng)
```

```python
sns.histplot(x='bill_length_mm',
             data=peng,
             bins=20)
```

```python
sns.histplot(x='bill_length_mm',
             data=peng,
             bins=[30,40,50,55])
```

```python
sns.histplot(x='bill_length_mm',
             data=peng,
             binwidth=5,
             binrange=(30,60))
```

```python
sns.histplot(x='bill_length_mm',
             data=peng,
             stat='count'
            )
```

```python
sns.histplot(x='bill_length_mm',
             data=peng,
             stat='density'
            )
```

```python
sns.histplot(x='bill_length_mm',
             data=peng,
             hue='species'
            )
```

```python
sns.histplot(x='bill_length_mm',
             data=peng,
             hue='species',
             multiple='stack'
            )
```

```python
sns.histplot(x='species',
             data=peng,
             hue='sex',
             multiple='dodge',
             shrink=0.7
            )
```

```python
sns.histplot(x='bill_length_mm',
             data=peng,
             color='cyan',
            )
```

```python
sns.histplot(x='species',
             data=peng,
             hue='sex',
             multiple='dodge',
             shrink=0.7,
             palette='deep'
            )
```

```python
sns.histplot(x='bill_length_mm',
             data=peng,
             fill=False,
            )
```

# boxplot

```python
cars = sns.load_dataset('mpg')
```

```python
cars = cars.dropna()
```

```python
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
cars.head()
```

```python
sns.boxplot(cars.mpg)
```

```python
cars.mpg.describe()
```

```python
sns.boxplot(x=cars.origin,
            y=cars.mpg)
```

```python
sns.boxplot(x='origin',
            y='mpg',
           data=cars)
```

```python
sns.boxplot(x='origin',
            y='mpg',
            hue='cylinders',
           data=cars)
```

```python
sns.boxplot(x='origin',
            y='mpg',
           data=cars,
           order=['japan','europe','usa'])
```

```python
sns.boxplot(x='origin',
            y='mpg',
           data=cars,
           color='b')
```

```python
sns.boxplot(x='origin',
            y='mpg',
            width=1,
           data=cars)
```

```python
sns.boxplot(x='origin',
            y='mpg',
            linewidth=2.5,
           data=cars)
```

```python
sns.boxplot(x='origin',
            y='mpg',
            whis=3.5,
           data=cars)
```

```python
sns.boxplot(x='origin',
            y='mpg',
            fliersize=25,
           data=cars)
```

# violinplot

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x=cars.cylinders,
               y=cars.displacement)
```

```python
sns.violinplot(x='cylinders',
               y='displacement',
               hue='origin',
              data=cars[cars.origin.isin(['japan','usa'])],
              split=True,
              inner='quartiles')
plt.legend(loc=2)
```

```python
%autosave 10
```

```python
sns.violinplot(x='cylinders',
               y='displacement',
               hue='origin',
              data=cars[cars.origin.isin(['japan','usa'])],
              split=True,
              scale='count',
              scale_hue=False)
plt.legend(loc=2)
```
