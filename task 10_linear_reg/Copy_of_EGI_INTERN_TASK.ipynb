```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Calories', axis=1) 
y = df['Calories']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=0)
```

```python
import matplotlib.pyplot as plt
for col in X_train.columns:
    plt.scatter(X_train[col], y_train, label='Train set', color='r', alpha=0.7)
    plt.scatter(X_test[col], y_test, label='Test set', color='b', alpha=0.7)
    plt.xlabel(col)
    plt.ylabel('Calories')
    plt.legend()
    plt.show()
```

### - Linear Regression Model
- Create linear model and train it
- Use linear regression model to predict on test data
- Plot prediction line against actual test data
- What is tge mean absolute error for this model?


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(X_train,y_train)
```

```python
y_pred_linear = linear_model.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred_linear)
```

##  Ridge Model
- Create ridge model and train it
- Use ridge model to predict on test data
- Plot prediction line against actual test data
- What is the mean absolute error for this model?


```python
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha=1).fit(X_train, y_train)
```

```python
y_pred_ridge = ridge_model.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred_ridge)
```

## Lasao model
- Create lasso model and train it
- Use lasso model to predict on test data
- Plot prediction line against actual test data
- What is tge mean absolute error for this model?


```python
from sklearn.linear_model import Lasso
lasso_model = Lasso().fit(X_train, y_train)
```

```python
y_pred_lasso  = lasso_model.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred_lasso)
```
